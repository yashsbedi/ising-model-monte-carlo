```python
import numpy as np

L = np.array([16, 24, 32, 40, 48, 56, 64], dtype=float)
Cv_max = np.array([1.479, 1.844, 1.838, 2.135, 1.905, 3.851, 2.425])
chi_max = np.array([57.955, 68.499, 132.944, 101.774, 198.101, 392.904, 469.802])

# Choose the sizes :
mask = (L >= 24) & (L != 56)   # 24, 32, 40, 48, 64

L_fit = L[mask]
Cv_fit = Cv_max[mask]
chi_fit = chi_max[mask]

logL = np.log(L_fit)
logCv = np.log(Cv_fit)
logChi = np.log(chi_fit)

# Linear fits in logâ€“log space
alpha_over_nu, Cv_intercept = np.polyfit(logL, logCv, 1)
gamma_over_nu, chi_intercept = np.polyfit(logL, logChi, 1)

print(alpha_over_nu, gamma_over_nu)



import matplotlib.pyplot as plt

plt.plot(np.log(L_fit), np.log(chi_fit), "o", label="data")
plt.plot(np.log(L_fit), gamma_over_nu * np.log(L_fit) + chi_intercept,
         label=f"fit, slope = {gamma_over_nu:.3f}")
plt.xlabel("log L")
plt.ylabel("log chi_max(L)")
plt.legend()
plt.show()
```
